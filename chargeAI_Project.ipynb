```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df =pd.read_csv("companies.csv")
```

```python
df
```

```python
x= df[["Productivity_Exp","Management_Exp","Promotions_Exp"]]
```

```python
x
```

```python
y = df[["Profit"]]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test , y_train ,y_test =train_test_split(x,y,test_size = 0.2 , random_state =True)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
```

```python
model1
```

```python
model1.fit(x_train,y_train)
```

```python
model1.intercept_
```

```python
model1.coef_
```

```python
model1.predict([[99069.95,151283.14,247029.42]])
```

```python
y_pred = model1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error
r2_score(y_test,y_pred)
```

```python
dataset = pd.DataFrame({"Prod_exp":x_test.iloc[:,0]
                        ,"Management_Exp":x_test.iloc[:,1],	"Promotions_Exp":x_test.iloc[:,2]	,"profit":y_test.squeeze(),"Predicted profit":y_pred.squeeze(),"Difference":(y_test-y_pred).squeeze()})
dataset
```
